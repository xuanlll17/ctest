```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

lastest_df = pd.read_csv("../six_edu_2023.csv")


levels = ['產業別', '地區', '教育程度類別']  # Modify the levels according to your DataFrame
color_columns = ['信用卡交易金額', '信用卡交易筆數']
value_column = '信用卡交易筆數'

def build_hierarchical_dataframe(df, levels, value_column, color_columns=None):
    df_all_trees = pd.DataFrame(columns=['id', 'parent', 'value', 'color'])
    for i, level in enumerate(levels):
        df_tree = pd.DataFrame(columns=['id', 'parent', 'value', 'color'])
        dfg = df.groupby(levels[i:]).sum()
        dfg = dfg.reset_index()
        df_tree['id'] = dfg[level].copy()
        if i < len(levels) - 1:
            df_tree['parent'] = dfg[levels[i+1]].copy()
        else:
            df_tree['parent'] = 'total'
        df_tree['value'] = dfg[value_column]
        df_tree['color'] = dfg[color_columns[0]] / dfg[color_columns[1]]
        
        if not df_all_trees.empty:
            df_all_trees = pd.concat([df_all_trees, df_tree], ignore_index=True)
        else:
            df_all_trees = df_tree
    
    total = pd.Series(dict(id='total', parent='',
                              value=df[value_column].sum(),
                              color=df[color_columns[0]].sum() / df[color_columns[1]].sum()))
    
    if not df_all_trees.empty:
        df_all_trees = pd.concat([df_all_trees, pd.DataFrame([total])], ignore_index=True)
    else:
        df_all_trees = pd.DataFrame([total])
    
    return df_all_trees


df_all_trees = build_hierarchical_dataframe(lastest_df, levels, value_column, color_columns)
average_score = lastest_df['信用卡交易金額'].sum() / lastest_df['信用卡交易筆數'].sum()

fig = make_subplots(1, 2, specs=[[{"type": "domain"}, {"type": "domain"}]],)

fig.add_trace(go.Sunburst(
    labels=df_all_trees['id'],
    parents=df_all_trees['parent'],
    values=df_all_trees['value'],
    branchvalues='total',
    marker=dict(
        colors=df_all_trees['color'],
        colorscale='RdBu',
        cmid=average_score),
    hovertemplate='<b>%{label} </b> <br> 信用卡交易筆數: %{value}<br> 信用卡交易金額: %{color:.2f}',
    name=''
    ), 1, 1)

fig.add_trace(go.Sunburst(
    labels=df_all_trees['id'],
    parents=df_all_trees['parent'],
    values=df_all_trees['value'],
    branchvalues='total',
    marker=dict(
        colors=df_all_trees['color'],
        colorscale='RdBu',
        cmid=average_score),
    hovertemplate='<b>%{label} </b> <br> 信用卡交易筆數: %{value}<br> 信用卡交易金額: %{color:.2f}',
    maxdepth=2
    ), 1, 2)

fig.update_layout(margin=dict(t=10, b=10, r=10, l=10))
fig.show()
```

```python
from dash import Dash, html, dcc, callback, Input, Output, State
import base64
import os

image_folder = 'assets/images/'

def b64_image(image_filename):
    with open(image_filename, 'rb') as f:
        image = f.read()
    return 'data:image/png;base64,' + base64.b64encode(image).decode('utf-8')

image_paths = [os.path.join(image_folder, filename) for filename in os.listdir(image_folder)]

app = Dash(__name__)

app.layout = html.Div([
    html.H1('Dash'),
    
    html.Button('Show Images', id='show-button'),
    
    html.Div(id='image-container'),
    
    dcc.Store(id='button-state', data=False)
])

@app.callback(
    Output('image-container', 'children'),
    Input('show-button', 'n_clicks'),
    State('button-state', 'data')
)
def update_images(n_clicks, button_state):
    if n_clicks is None:
        return []
    
    # Toggle button state
    button_state = not button_state
    
    # Return images only if the button is clicked
    if button_state:
        return [html.Img(src=b64_image(image_path)) for image_path in image_paths]
    else:
        return []

@app.callback(
    Output('button-state', 'data'),
    Input('show-button', 'n_clicks')
)
def update_button_state(n_clicks):
    if n_clicks is None:
        return False
    
    # Toggle button state
    return not n_clicks % 2 == 0

if __name__ == "__main__":
    app.run_server(host="localhost", port=8050)
```

```python
from dash import Dash, html, dcc, callback, Input, Output, State
import base64
import os

image_folder = 'assets/images/'

def b64_image(image_filename):
    with open(image_filename, 'rb') as f:
        image = f.read()
    return 'data:image/png;base64,' + base64.b64encode(image).decode('utf-8')

image_paths = [os.path.join(image_folder, filename) for filename in os.listdir(image_folder)]

app = Dash(__name__)

app.layout = html.Div([
    html.H1('Dash'),
    
    html.Button('Show Image', id='show-button'),
    
    html.Img(id='displayed-image'),
    
    dcc.Store(id='image-index', data=0)
])

@app.callback(
    Output('displayed-image', 'src'),
    Input('show-button', 'n_clicks'),
    State('image-index', 'data')
)
def update_image_src(n_clicks, image_index):
    if n_clicks is None:
        return b64_image(image_paths[0])
    
    # Increment image index and reset if it exceeds the number of images
    image_index = (image_index + 1) % len(image_paths)
    
    return b64_image(image_paths[image_index])

@app.callback(
    Output('image-index', 'data'),
    Input('show-button', 'n_clicks')
)
def update_image_index(n_clicks):
    if n_clicks is None:
        return 0
    
    return n_clicks % len(image_paths)

if __name__ == "__main__":
    app.run_server(host="localhost", port=8050)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

lastest_df = pd.read_csv('../six_edu_2023.csv')

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='month',
        options=[
            {'label': '一月', 'value': '1'},
            # Add other months as needed
        ],
        value='ALL',
        style={'width': '50%'}
    ),
    dcc.Graph(id='graph_sunburst'),
])

@app.callback(
    Output("graph_sunburst", "figure"),
    Input("month", "value")
)
def update_sunburst_chart(selected_mon):
    global lastest_df
    if selected_mon == "ALL":
        fig = px.sunburst(lastest_df, path=['年', '月', '產業別', '教育程度類別'], values='信用卡交易金額',width=500, height=600)
    else:
        filtered_df = lastest_df[lastest_df['月'].astype(str) == selected_mon]
        fig = px.sunburst(filtered_df, path=['月', '產業別', '教育程度類別'], values='信用卡交易金額',width=1800, height=800)
    return fig

if __name__ == '__main__':
    app.run_server(host="localhost", port=8050,debug=True)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Assuming df is your DataFrame with the desired data
dataset = pd.read_csv("../age.csv")
df = pd.DataFrame(dataset)
df['平均交易金額'] = df['信用卡交易金額[新台幣]'] / df['信用卡交易筆數']

# Aggregate data by summing values for each industry
agg_df = df.groupby('產業別').agg({
    '信用卡交易金額[新台幣]': 'sum',
    '信用卡交易筆數': 'sum',
    '平均交易金額': 'mean'  # You can adjust this based on your specific requirement for average
}).reset_index()

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add bar trace
fig.add_trace(
    go.Bar(x=agg_df['產業別'], y=agg_df['信用卡交易金額[新台幣]'], name="信用卡交易金額[新台幣]"),
    secondary_y=False,
)

# Add line trace
fig.add_trace(
    go.Scatter(x=agg_df['產業別'], y=agg_df['平均交易金額'], name="平均交易金額", mode='lines+markers', marker=dict(color='red')),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="不同產業別的信用卡交易金額及平均交易金額"
)

# Set x-axis title
fig.update_xaxes(title_text="產業別")

# Set y-axes titles
fig.update_yaxes(title_text="信用卡交易金額[新台幣]", secondary_y=False)
fig.update_yaxes(title_text="平均交易金額", secondary_y=True)

# Show the plot
fig.show()
```
