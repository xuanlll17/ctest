```python
import plotly.express as px
import pandas as pd

# 讀取 CSV 檔案
df = pd.read_csv("../six_age_2023.csv")

# 將資料按照年和月進行分組，計算每個月的信用卡消費金額總和
monthly_total = df.groupby(['年', '月'])['信用卡交易金額'].sum().reset_index()

# 繪製折線圖
fig = px.line(monthly_total, x="月", y="信用卡交易金額", color="年", title='每月信用卡消費金額變化', markers=True)

# 顯示圖表
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# 讀取 CSV 檔案
df = pd.read_csv("../six_age_2023.csv")

# 條件篩選：只選取每年的一月
january_data = df[(df['月'] == 1)]
display(january_data)

# 繪製折線圖
fig = px.bar(january_data, x="地區", y="信用卡交易金額", facet_col="月", title='每年一月信用卡消費金額變化')

# 顯示圖表
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# 讀取 CSV 檔案
df = pd.read_csv("../six_age_2023.csv")

# 選擇產業別為 '衣' 的資料
filtered_df = df[df['產業別'] == '衣']
display(filtered_df)

# 使用 px.pie 繪製圓餅圖
fig = px.pie(filtered_df, values='信用卡交易金額', names='年齡層',
             title='各年齡層在衣產業別的信用卡消費總金額')

# 顯示圖表
fig.show()
```
